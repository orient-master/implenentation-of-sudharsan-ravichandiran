```python
def sigmoid(x):
    return 1/ (1+np.exp(-x))

x = np.linspace(-10, 10, 400)  # Generates 400 points between -10 and 10

# Calculate y values using the sigmoid function
y = sigmoid(x)

# Plotting the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function', color='b')
plt.title('Sigmoid Function Plot')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def tanh(x):
    numerator = 1-np.exp(-2*x)
    denominator = 1+np.exp(-2*x)
    return numerator/denominator

# Generate x values
x = np.linspace(-10, 10, 400)  # Generates 400 points between -10 and 10

# Calculate y values using the tanh function
y = tanh(x)

# Plotting the tanh function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Tanh Function', color='r')
plt.title('Tanh Function Plot')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the ReLU function
def ReLU(x):
    return np.maximum(0, x)  # Vectorized version of ReLU for arrays

# Generate x values
x = np.linspace(-10, 10, 400)  # Generates 400 points between -10 and 10

# Calculate y values using the ReLU function
y = ReLU(x)

# Plotting the ReLU function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='ReLU Function', color='g')
plt.title('ReLU Function Plot')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the Leaky ReLU function
def leakyReLU(x, alpha=0.01):
    return np.where(x < 0, alpha * x, x)  # Vectorized version using np.where

# Generate x values
x = np.linspace(-10, 10, 400)  # Generates 400 points between -10 and 10

# Calculate y values using the Leaky ReLU function
y = leakyReLU(x, alpha=0.01)

# Plotting the Leaky ReLU function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Leaky ReLU Function', color='m')
plt.title('Leaky ReLU Function Plot')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def ELU(x,alpha=0.01):
    if x<0:
        return ((alpha*(np.exp(x)-1)))
    else:
        return x
```

```python
# Define the ELU function
def ELU(x, alpha=0.01):
    return np.where(x < 0, alpha * (np.exp(x) - 1), x)  # Vectorized version using np.where

# Generate x values
x = np.linspace(-10, 10, 400)  # Generates 400 points between -10 and 10

# Calculate y values using the ELU function
y = ELU(x, alpha=0.01)

# Plotting the ELU function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='ELU Function', color='c')
plt.title('ELU Function Plot')
plt.xlabel('x')
plt.ylabel('ELU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def swish(x,beta):
    return 2*x*sigmoid(beta*x)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the Swish function
def swish(x, beta):
    return x * sigmoid(beta * x)

# Generate x values
x = np.linspace(-10, 10, 400)  # Generates 400 points between -10 and 10

# Define the beta parameter
beta = 1  

# Calculate y values using the Swish function
y = swish(x, beta)

# Plotting the Swish function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Swish Function', color='b')
plt.title('Swish Function Plot')
plt.xlabel('x')
plt.ylabel('Swish(x)')
plt.grid(True)
```

```python
def softmax(x):
    return np.exp(x) / np.exp(x).sum(axis=0)
```

```python
# Define the Softmax function
def softmax(x):
    e_x = np.exp(x - np.max(x))  # Subtracting max(x) for numerical stability
    return e_x / e_x.sum(axis=0)

# Generate x values (logits)
x = np.linspace(-10, 10, 400)  # Generates 400 points between -10 and 10

# Calculate y values using the Softmax function
logits = np.stack([x, x + 1, x + 2], axis=0)
y = softmax(logits)  # Apply softmax to the logits

# Plotting the Softmax function
plt.figure(figsize=(10, 6))
for i in range(y.shape[0]):
    plt.plot(x, y[i], label=f'Class {i+1}')

plt.title('Softmax Function Plot')
plt.xlabel('x')
plt.ylabel('Softmax(x)')
plt.legend()
plt.grid(True)
plt.show()
```
